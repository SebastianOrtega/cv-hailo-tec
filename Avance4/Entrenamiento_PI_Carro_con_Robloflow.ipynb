```python
from ultralytics import YOLO
import cv2

# Carga del modelo YOLO
model = YOLO('/content/runs/detect/train3/weights/best.pt')

# ConfiguraciÃ³n del path del video y del modelo
video_path = '/content/runs/detect/train3/Corto.mp4'

# ConfiguraciÃ³n para guardar el video con detecciones
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('/content/detected_video.mp4', fourcc, fps, (width, height))

# Procesar el video en modo stream
results = model.predict(video_path, imgsz=640, conf=0.7, save=False, show=False, stream=True)

for result in results:
    # Procesar cada frame
    frame = result.render()[0]  # Renderiza las detecciones en el frame

    # Guardar el frame en el video de salida
    out.write(frame)

    # Opcional: Mostrar el frame (si estÃ¡s trabajando localmente y quieres verlo en tiempo real)
    # cv2.imshow('Frame', frame)
    # if cv2.waitKey(1) & 0xFF == ord('q'):
    #     break

# Liberar recursos
cap.release()
out.release()
cv2.destroyAllWindows()
```
